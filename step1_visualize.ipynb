```python
# Import the RDKit chemistry modules
from rdkit import Chem
from rdkit.Chem import Draw

# The SMILES string for Ibuprofen
smiles_str = "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"

# Convert the string into an RDKit molecule object
molecule = Chem.MolFromSmiles(smiles_str)

# Display the 2D structure right here in the notebook
Draw.MolToImage(molecule, size=(400, 400))
```
